```python
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("Churn_Modelling.csv")
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.shape
```

```python
X = df.iloc[:,3:13].values
y = df.iloc[:,13].values
```

```python
# labelencoding and onehotencoding
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
X_1 = LabelEncoder()
X[:,1] = X_1.fit_transform(X[:,1])
X_2 = LabelEncoder()
X[:,2] = X_2.fit_transform(X[:,2])
```

```python
pd.DataFrame(X)
```

```python
onehotencoder = OneHotEncoder(categorical_features=[1])
X = onehotencoder.fit_transform(X).toarray()
```

```python
X =X[:,1:]
```

```python
#standarscalar
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
X = scalar.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2 , random_state=101)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
classifier = Sequential()
classifier.add(Dense(units = 6 , activation = "relu" , kernel_initializer = 'normal' , input_dim = 11))
classifier.add(Dense(units = 6 , activation = "relu" , kernel_initializer = 'normal' ))
classifier.add(Dense(units = 1 , activation = "sigmoid" , kernel_initializer = 'normal' ))
```

```python
classifier.compile(optimizer="adam",loss = 'binary_crossentropy' , metrics=['accuracy'])
```

```python
classifier.fit(X_train , y_train , epochs=100 , batch_size=10)
```

```python
#predicting the unknown values values
y_pred = classifier.predict(scalar.transform(np.array([[0,0, 600, 1, 40,3,60000,2,1,1,50000]])))
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix
confusion_matrix(y_pred , y_test)
```

```python
accuracy_score(y_pred , y_test)
```

```python
#evaluating using cross_val_score
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score

def classifier_model():
    model1 = Sequential()
    model1.add(Dense(units = 6 , activation = "relu" , kernel_initializer = 'normal' , input_dim = 11))
    model1.add(Dense(units = 6 , activation = "relu" , kernel_initializer = 'normal'))
    model1.add(Dense(units = 1 , activation = "sigmoid" , kernel_initializer = 'normal'))
    model1.compile(optimizer="adam" , loss= 'binary_crossentropy' , metrics=['accuracy'])
    return model1

model1 = KerasClassifier(build_fn=classifier_model, epochs = 100 , batch_size = 10)
accuracy = cross_val_score(estimator = model1 , X= X_train , y =y_train , cv =10)
```

```python
 accuracy.mean()
```

```python
accuracy.std()
```

```python
# dropout regularization to reduce overfitting
def drop_model():
    model2 = Sequential()
    model2.add(Dense(units= 6 , activation = 'relu' , input_dim = 11 , kernel_initializer = 'normal' ))
    model2.add(Dropout(p=0.1))
    model2.add(Dense(units= 6 , activation = 'relu' ,  kernel_initializer = 'normal' ))
    model2.add(Dropout(p=0.1))
    model2.add(Dense(units= 1 , activation = 'sigmoid' , kernel_initializer = 'normal' ))
    model2.compile(optimizer='adam' , loss = "binary_crossentropy" , metrics=['accuracy'])
    return model2

model2 = KerasClassifier(build_fn = drop_model , epochs = 100 , batch_size = 10)
accuracy = cross_val_score(estimator=model2 , X= X_train , y = y_train , cv =10)
```

```python
accuracy.mean()
```

```python
# parameter tuning using GridSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
%%timeit
def tuning_model():
    model3 = Sequential()
    model3.add(Dense(units=6 , activation = 'relu' , input_dim = 11 , kernel_initializer = 'normal'))
    model3.add(Dense(units=6 , activation = 'relu'  , kernel_initializer = 'normal'))
    model3.add(Dense(units=1 , activation = 'sigmoid' ,  kernel_initializer = 'normal'))
    model3.compile(optimizer= 'adam' , loss = 'binary_crossentropy', metrics = ['accuracy'])
    return model3

model3 = KerasClassifier(build_fn=tuning_model)
parameter = {
            'epochs' : [100,200],
            'batch_size': [25,30]
            }

gridsearch = GridSearchCV(estimator = model3 , param_grid= parameter , scoring= 'accuracy' , cv=10)
gridsearch.fit(X_train,y_train)
```
